```python
import matplotlib.pyplot as plt
import numpy as np


# Prepare data
predicted_values = [1.7118, 1.7891, 1.8909, 2.0255,2.0925, 2.1682]
real_values = [1.7, 1.8, 1.9, 2, 2.1, 2.2]
x_axis = np.arange(len(predicted_values))  # x-axis ticks

# Calculate absolute relative errors
relative_errors = [abs((pred - real)/real) for pred, real in zip(predicted_values, real_values)]

# First plot: Predicted vs Real values
plt.figure(figsize=(10, 6))
plt.plot(x_axis, predicted_values, 'o-', color='blue', label='Predicted values of v by PINN')
plt.plot(x_axis, real_values, 's--', color='red', label='Real Values of v')



# Set titles and labels
plt.title('Predicted Values of v by PINN vs Real Values of v', fontsize=14)
plt.xlabel('Data Point Index', fontsize=12)
plt.ylabel('v Values', fontsize=12)
plt.xticks(x_axis, [f' {i+1}' for i in x_axis])
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Second plot: Relative Errors
plt.figure(figsize=(10, 6))
plt.plot(real_values, relative_errors, 'D-', color='green', label='Relative Error')

# Add error labels
for v, error in zip(real_values, relative_errors):
    plt.text(v, error, f'{error:.2%}', ha='center', va='bottom')

# Set titles and labels
plt.title('Relative Errors for Different v Values', fontsize=14)
plt.xlabel('Real values of v', fontsize=12)
plt.ylabel('Relative Errors', fontsize=12)
plt.xticks(real_values)  # Use actual v values as x-ticks
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
